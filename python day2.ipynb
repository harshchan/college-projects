```python
#recursive functions

def recsum(sum):
    a=int(input("\n Enter number less than 10 "))
    if(a<10):
        sum=a+sum
        print("sum is ",sum)
        recsum(sum)
    else:
        return 0
    
recsum(0)
    
```
